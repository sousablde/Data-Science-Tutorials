```python
#reading the data
df_can = pd.read_excel('https://ibm.box.com/shared/static/lw190pt9zpy5bd1ptyg2aw15awomz9pu.xlsx',
                       sheetname='Canada by Citizenship',
                       skiprows=range(20),
                       skip_footer=2)

print ('Data read into a pandas dataframe!')
```

```python
#cleaning the data set
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
```

```python
#rename the columns
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

```python
#add column for total immmigrants
df_can['Total'] = df_can.sum(axis=1)
```

```python
#check for null objects
df_can.isnull().sum()
```

```python
#select specific row corresponding to a country
df_can.set_index('Country', inplace=True)
df_can.head(3)
```

```python
#convert the column names into strings
df_can.columns = list(map(str, df_can.columns))
# [print (type(x)) for x in df_can.columns.values] #<-- uncomment to check type of column headers
```

```python
#declare a variable that will allow us to easily call upon the full range of years
years = list(map(str, range(1980, 2014)))
```

```python
#filter the dataframe to show the data on Asian countries (AreaName = Asia)
# 1. create the condition boolean series
condition = df_can['Continent']=='Asia'
print (condition)

# 2. pass this condition into the dataFrame
df_can[condition]

# we can pass mutliple criteria in the same line. 
# let's filter for AreaNAme = Asia and RegName = Southern Asia

df_can[(df_can['Continent']=='Asia') & (df_can['Region']=='Southern Asia')]
```

```python
#importing Matplotlib and Matplotlib.pyplot
# we are using the inline backend
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#apply a style to Matplotlib.
print(plt.style.available)
mpl.style.use(['ggplot']) # optional: for ggplot-like style
```

```python
#extract the data series for Haiti
haiti = df_can.loc['Haiti', years] # Passing in years 1980 - 2013 to exclude the 'total' column
haiti.head()
haiti.plot()
```

```python
#label the x and y axis using plt.title(), plt.ylabel(), and plt.xlabel()
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show() # Need this line to show the updates made to the figure
```

```python
#annotate this spike in the plot by using the plt.text() method regarding earthquake
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake. 
# syntax: plt.text(x, y, label)
plt.text(20, 6000, '2010 Earthquake') # see note below

plt.show() 
```

```python
#compare india and china
#get the dataset
df_CI = df_can.loc[['India', 'China'], years]
df_CI.head()

#plot graph
df_CI.plot(kind='line')

#transpose columns and rows
df_CI = df_CI.transpose()
df_CI.head()

#pass a title parameter to the plot() method as an alternate approach to add the title
df_CI.plot(kind='line')

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show() 
```

```python
#Compare the trend of top 5 countries that contributed the most to immigration to Canada
#Get the dataset. Recall that we created a Total column that calculates the cumulative immigration by country. We will sort on this column to get our top 5 countries using pandas sort_values() method
# inplace = True paramemter saves the changes to the original df_can dataframe
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

# get the top 5 entries
df_top5 = df_can.head(5)

# transpose the dataframe
df_top5 = df_top5[years].transpose() 

df_top5

#Plot the dataframe
df_top5.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```
